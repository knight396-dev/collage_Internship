3) Import an meaningful csv file for data analysis and perform data cleaning, and analysis for meaningful output

```python
df = pd.read_csv("sales_data.csv")

print("\nOriginal Data:")
print(df)

# Data Cleaning
df["Date"] = pd.to_datetime(df["Date"])
df["Product"] = df["Product"].str.strip()
df["Quantity"] = df["Quantity"].fillna(0)
df["Price"] = df["Price"].fillna(df["Price"].mean())

df["Total_Sales"] = df["Quantity"] * df["Price"]

print("\nCleaned Data:")
print(df)

print("\nTotal Sales:")
print(df["Total_Sales"].sum())

print("\nSales by Product:")
print(df.groupby("Product")["Total_Sales"].sum())

print("\nAverage Price:")
print(df["Price"].mean())

print("\nHighest Sales Row:")
print(df.loc[df["Total_Sales"].idxmax()])
```
