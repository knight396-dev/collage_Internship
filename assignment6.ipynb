```python
#game : pixel drawing using turtle module

'''


recomenddation   : run it on python idle insted od jupyter notebook because it is not working on jupyter notebook
 for better experince run it on python idle insted od jupyter notebook because it is not working on jupyter notebook

'''


import turtle

screen = turtle.Screen() # to greamt a game window 

#to set size ofthe window
screen.setup(width=800, height=600)


# player setup 
player = turtle.Turtle()
player.shape("turtle")
player.color("blue")
#player.penup()  # donesent draw the line 
player. goto(0, -250) # initalization

def move_left():
    x = player.xcor() # gets current position
    x = x - 20
    player.setx(x) # change posrtion 

def move_right():
    x = player.xcor()
    x = x + 20
    player.setx(x)

def move_up():
    y = player.ycor() # gets current position
    y = y + 20
    player.sety(y) # change posrtion 

def move_down():
    y = player.ycor()
    y = y - 20
    player.sety(y)


# prom;lem two : turtuel usle Left insted of left
screen.listen()
screen.onkeypress(move_left,"Left")
screen.onkeypress(move_right,"Right")
screen.onkeypress(move_up,"Up")
screen.onkeypress(move_down,"Down")



screen.mainloop() # NOTHING RUNS AFTER THAT

```
